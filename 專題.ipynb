# 文字雲

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

font_path = "NotoSansTC-SemiBold.ttf"

# 要生成文字雲的文字及其出現的次數
word_frequencies = {
    "微積分": 93.5,
    "程式設計": 98.5,
    "中級會計學": 99,
    "金融科技": 96,
    "投資學": 95,
    "統計": 91,
    "資料庫管理": 92,
    "商業程式設計": 99,
    "大數據分析": 99, 
    "金融市場": 98, 
    "系統分析與設計": 97, 
    "財務報表分析": 99, 
    "人工智慧概論": 99, 
    "人工智慧跨域商情預測之應用": 98, 
    "AI機器視覺結合深度學習實作": 100, 
    "企業倫理": 100, 
    "人工智慧跨域金融之應用": 90, 
    "管理資訊系統": 95, 
    "企業資料通訊": 94, 
    "財務金融實習": 94, 
    "管理學": 98,
    "財務管理": 100,
    "財金時事選讀": 100
}

# 生成文字雲
wordcloud = WordCloud(font_path = font_path, width = 800, height = 400, background_color = 'white').generate_from_frequencies(word_frequencies)

# 顯示文字雲
plt.figure(figsize = (10, 5))
plt.imshow(wordcloud, interpolation = "lanczos") # 改變圖像的大小或解析度，以平滑圖像並減少鋸齒狀邊緣效應。
plt.axis("off")
# plt.tight_layout(pad = 0)
plt.show()
```
